```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=np.arange(0,10,2,dtype="int")
```

```python
x
```

```python
y=2*x
```

```python
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line Plot")
plt.xlim(0)
plt.ylim(0)
plt.savefig("Firstplot.png");
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1]);
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y);
```
