# Data Preprocessing

```python
df.isnull().sum()
```

```python
print(df["Rating"].min())
print(df["Rating"].max())
```

```python
df["Rating"]=df["Rating"].fillna(df["Rating"].mean())
```

```python
print(df["Size"].min())
print(df["Size"].max())
```

```python
df["Size"]=df["Size"].fillna(df["Size"].mean())
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
import re
def remove_special_chars(text):
    return re.sub(r'[^a-zA-Z0-9]', '', text)
df["Installs"]= df["Installs"].apply(lambda x: remove_special_chars(x))
df["Price"]= df["Price"].apply(lambda x: remove_special_chars(x))
```

```python
df.head()
```

```python
df["Installs"]= df["Installs"].astype("float")
df["Price"]= df["Price"].astype("float")
```

```python
df.describe().T
```

# Sentiment Analysis Summary

```python
sentiment_summary = df_fin.groupby('Sentiment').agg({'Sentiment_Polarity': 'mean', 'Sentiment_Subjectivity': 'mean'})
print(sentiment_summary)
```

# Visualize Sentiment Analysis Summary

```python
plt.figure(figsize=(10, 6))
sentiment_summary.plot(kind='bar')
plt.title('Sentiment Analysis Summary')
plt.xlabel('Sentiment')
plt.ylabel('Mean Value')
plt.show()
```

# Sentiment Analysis

```python
from textblob import TextBlob
def analyze_sentiment(text):
    analysis = TextBlob(str(text))
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
df_fin['Sentiment_TextBlob'] = df_fin['Translated_Review'].apply(analyze_sentiment)
print(df_fin[['App', 'Translated_Review', 'Sentiment_TextBlob']].head())
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Sentiment_TextBlob', data=df_fin)
plt.title('Sentiment Analysis by TextBlob')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```
